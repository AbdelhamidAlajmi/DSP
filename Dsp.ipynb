```python
import numpy as np
import matplotlib.pyplot as plt

# Function to evaluate the signal based on the given formula
def evaluate_signal(formula, x):
    # Replace the variable 't' with the input values
    expr = formula.replace('t', str(x))
    # Evaluate the expression using 'eval' function
    return eval(expr)

# Get the input formula from the user
formula = input("Enter the formula for the signal: ")

# Define the range of values for the independent variable (t)
t_start = 0
t_end = 10
num_points = 1000
t = np.linspace(t_start, t_end, num_points)

# Evaluate the signal for each value of t
signal = [evaluate_signal(formula, x) for x in t]

# Plot the signal
plt.plot(t, signal)
plt.xlabel('t')
plt.ylabel('Signal')
plt.title('Signal Shape')
plt.grid(True)
plt.show()
```
